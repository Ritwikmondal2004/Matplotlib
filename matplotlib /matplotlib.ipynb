```python
import matplotlib.pyplot as plt
import numpy as np
```

## first create a graph

```python
plt.plot([1, 2, 3, 4])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show() #for show the graph
```

## create 2nd graph

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show() ## for using 
```

## showing x and y 

```python
x=[1,2,3]
y=[5,6,7]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show() ## for using 
```

## marks calculation

```python
x=[1,2,3,4]
y=[90,67,92,96]
plt.plot(x,y)
plt.title("marks for school") ## for head line
plt.xlabel('student-id')
plt.ylabel('marks')
plt.show() ## for using 
```

## use of x-ticks and y-ticks

```python
x=[1,2,3,4]
y=[93,95,96,98]
plt.plot(x,y)
plt.title("marks for school") ## for head line
plt.xlabel('student-id')
plt.ylabel('marks')
plt.xticks([1,2,3,4])
plt.yticks([98,92,94,96,98,100])
plt.show() ## for using 
```

## double map

```python
x=[1,2,3,4]
dsa=[93,95,96,98]
ml=[95,97,98,100]
plt.plot(x,dsa)
plt.plot(x,ml)
plt.title("marks for school") ## for head line
plt.xlabel('student-id')
plt.ylabel('marks')
plt.xticks([1,2,3,4])
plt.yticks([98,92,94,96,98,100])
plt.show() ## for using 
```

## using label

```python
x = [1, 2, 3, 4]
dsa = [93, 95, 96, 98]
ml = [95, 97, 98, 100]

plt.plot(x, dsa, label='dsa marks',color="blue",linewidth=2,marker='^') ## using colour to change (colour)** linewidth=2,marker='*' linestyle='---' you can use 
plt.plot(x, ml, label='ml marks')
plt.title("marks for school")  # Headline
plt.xlabel('student-id')
plt.ylabel('marks')
plt.xticks([1, 2, 3, 4])
plt.yticks([92, 94, 96, 98, 100])  # Removed duplicate 98
plt.legend()  # Add this to show the labels   using (loc='lower right') to position of this
plt.show()
```

## save any image

```python
x = [1, 2, 3, 4]
dsa = [93, 95, 96, 98]
ml = [95, 97, 98, 100]

plt.plot(x, dsa, label='dsa marks')
plt.plot(x, ml, label='ml marks')
plt.title("marks for school")  # Headline
plt.xlabel('student-id')
plt.ylabel('marks')
plt.xticks([1, 2, 3, 4])
plt.yticks([92, 94, 96, 98, 100])  # Removed duplicate 98
plt.legend()  # Add this to show the labels
plt.show()


plt.savefig('marks.jpg',dpi=1000)
```
